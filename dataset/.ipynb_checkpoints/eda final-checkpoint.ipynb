```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loan_df=pd.read_csv("Clean Lending Club Loans")
info=pd.read_csv('LCDataDictionary.csv')
```

```python
# Creating Function to get column name explanation
def search(x):
    fnd=info[info['LoanStatNew']==str(x)].values
    print('column name is ---->',fnd[0][0])
    print('Description is ---->',fnd[0][1])
```

```python
loan_df.info()
```

```python
col_to_dtime=['issue_d','last_pymnt_d','last_credit_pull_d','earliest_cr_line']
for i in col_to_dtime:
    loan_df[i]=pd.to_datetime(loan_df[str(i)])
```

```python
loan_df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
for i in loan_df.columns:
    print(search(i))
    print('-----------------------------------------------------------------------------------')
```

```python
search('grade')
```

```python
plt.figure(figsize=(14,5))
plt.suptitle("Number of borrowers in each grade assigned by Lending Club",fontsize=18,fontweight='bold')
sns.set_style('whitegrid')
plt.subplot(1,2,1)
sns.countplot(y='grade',data=loan_df)
plt.xlabel('Number of Borrowers',fontsize=12)
plt.ylabel("Grade's Assigned by Lending Club",fontsize=12)
plt.tight_layout()

plt.subplot(1,2,2)
plt.pie(loan_df['grade'].value_counts(),
        autopct='%.0f%%',
        labels=loan_df['grade'].unique())
plt.show()
```

```python
search('funded_amnt')
search('funded_amnt_inv')
```

```python
plt.figure(figsize=(18,6))
plt.suptitle('Total amount commietted to the loan by Lending Club Vs Investors at that point in each grade',fontsize=20,fontweight='bold')
plt.subplot(1,2,1)
plt.pie(loan_df.groupby('grade').mean().reset_index()['funded_amnt'],
        autopct='%.0f%%',
        labels=loan_df.groupby('grade').mean().reset_index()['grade'])
plt.title('Amount committed By Lending Club',fontsize=15,pad=-340)
plt.subplot(1,2,2)
plt.title('Amount committed By Lending Club Investors',fontsize=15,pad=-340)
plt.pie(loan_df.groupby('grade').mean().reset_index()['funded_amnt_inv'],
        autopct='%.0f%%',
        labels=loan_df.groupby('grade').mean().reset_index()['grade'])
plt.show()
```

```python
search('term')
```

```python
plt.figure(figsize=(14,6))
sns.countplot(x='grade',hue='term',data=loan_df,palette="mako")
plt.title("Number of loan payments in each grade",fontsize=18,fontweight='bold')
plt.xlabel("Grade's Assigned by Lending Club",fontsize=12)
plt.show()
```

```python
loan_df.groupby('grade').describe()['int_rate'].reset_index()
```

```python
plt.figure(figsize=(14,6))
plt.suptitle('Average intrest rate vs total intrest rate in each grade',fontsize=18,fontweight='bold')
plt.subplot(1,2,1)
sns.barplot(x='grade',y='mean',data=loan_df.groupby('grade').describe()['int_rate'].reset_index())
plt.subplot(1,2,2)
sns.barplot(x='grade',y='int_rate',data=loan_df.groupby('grade').sum()['int_rate'].reset_index())
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.boxplot(y='grade',x='int_rate',data=loan_df)
plt.xlabel('Intrest Rate')
plt.ylabel('Grade')

plt.title('Distribution of intrest rate in each grade',fontsize=18,fontweight='bold')
plt.show()
```
